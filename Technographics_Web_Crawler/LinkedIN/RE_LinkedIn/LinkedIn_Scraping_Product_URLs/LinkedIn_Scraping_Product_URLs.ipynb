```python
#Script to scrape product URLs and other information by visiting each linkedIn page using the Vendor hrefs we scraped along with Linkedln_Rescraped_Data.
os.environ['GH_TOKEN'] = ""

df_temp = pd.DataFrame()
templist = []

data = pd.read_csv("/Users/shagarwal/Documents/LinkedIN/RE_LinkedIn/LinkedIn_Scraping_Product_URLs/Input/LinkedIn_productURL_Input.csv")
for index,row in data[4:5].iterrows():
    Product_name = row[0]
    Product_href = row[1]
    #print(f"--------{parent_cat}, {sub_cat}, {href}------------")
    try:
        firefox_options = FirefoxOptions()
        #firefox_options.add_argument("--headless")
        driver = webdriver.Firefox(options=firefox_options, service=Service(GeckoDriverManager().install()))
        driver.maximize_window()
        driver.get(Product_href)
        driver.set_page_load_timeout(30)
        time.sleep(4)
        
        #
        try:
            URL_type = driver.find_element(By.XPATH, "//*[@id='main-content']/section/div/section/div/div/div/div/a[2]").text
        except:
            URL_type = " "
        
        #
        try:
            URL = driver.find_element(By.XPATH, "//*[@id='main-content']/section/div/section/div/div/div/div/a[2]").get_attribute("href")
        except:
            URL = " "
            
        #
        try:
            product_about_us = driver.find_element(By.XPATH, "//*[@id='main-content']/section/div/section/p").text
        except:
            product_about_us = " "
            
        #
        try:
            product_intended_for = driver.find_element(By.XPATH, "//*[@id='main-content']/section[1]/div/section[2]/ul").text
        except:
            product_intended_for = " "   
            
        print(f" {Product_name}, {URL_type} ")
        table_dict = {'Product_name': Product_name, 'Product_href': Product_href, 'URL_type': URL_type, 'URL': URL, 'product_about_us': product_about_us }
        templist.append(table_dict)
        #driver.quit()
        time.sleep(0.5)
    except Exception as e:
        print(e)
        time.sleep(10)
df_temp = pd.DataFrame(templist)
df_temp.to_csv(f"/Users/shagarwal/Documents/LinkedIN/RE_LinkedIn/LinkedIn_Scraping_Product_URLs/output/sample_linkedin_product_URLs.csv", index=False)
    
```
