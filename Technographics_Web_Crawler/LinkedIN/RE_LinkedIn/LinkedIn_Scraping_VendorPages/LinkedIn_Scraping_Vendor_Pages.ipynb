```python
#Script to scrape Vendor URLs and other information by visiting each linkedIn page using the Vendor hrefs we scraped along with Linkedln_Rescraped_Data.
os.environ['GH_TOKEN'] = ""

df_temp = pd.DataFrame()
templist = []

data = pd.read_csv("/Users/shagarwal/Documents/LinkedIN/RE_LinkedIn/LinkedIn_Scraping_ProductPages/input/unique_vendors.csv")
for index,row in data[3499:7000].iterrows():
    vendor_number = row[0]
    vendor_name = row[1]
    vendor_href = row[2]
 
    #print(f"--------{parent_cat}, {sub_cat}, {href}------------")
    try:
        firefox_options = FirefoxOptions()
        firefox_options.add_argument("--headless")
        driver = webdriver.Firefox(options=firefox_options, service=Service(GeckoDriverManager().install()))
        driver.maximize_window()
        driver.get(vendor_href)
        driver.set_page_load_timeout(30)
        time.sleep(4)
        
        #Close the login popup
        try:
            cross_button = driver.find_element(By.XPATH, "//*[@id='organization_guest_contextual-sign-in']/div/section/button/icon").click()
        except:
            print("Could not click on cross button")
        
        #About us - vendor
        try:
            about_us_vendor = driver.find_element(By.XPATH, "//*[@id='main-content']/section[1]/div/section[1]/div/p").text
        except:
            about_us_vendor = " "
            
        #website - vendor
        try:
            website_vendor = driver.find_element(By.XPATH, "//*[@id='main-content']/section[1]/div/section[1]/div/dl/div[1]/dd/a").text
        except:
            website_vendor = " "
            
        #Industries - vendor
        try:
            industry_vendor = driver.find_element(By.XPATH, "//*[@id='main-content']/section[1]/div/section[1]/div/dl/div[2]/dd").text
        except:
            industry_vendor = " "
        
        #Company Size -Vendor 
        try:
            company_size_vendor = driver.find_element(By.XPATH, "//*[@id='main-content']/section[1]/div/section[1]/div/dl/div[3]/dd").text
        except:
            company_size_vendor = " "
        
        #Headquarters - Vendor
        try:
            headquarters_vendor = driver.find_element(By.XPATH, "//*[@id='main-content']/section[1]/div/section[1]/div/dl/div[4]/dd").text
        except:
            headquarters_vendor = " "
        
        #Type - Vendor
        try:
            type_founded = driver.find_element(By.XPATH, "//*[@id='main-content']/section[1]/div/section[1]/div/dl/div[5]/dd").text
        except:
            type_founded = " "
            
        #Founded - Vendor
        try:
            founded_vendor = driver.find_element(By.XPATH, "//*[@id='main-content']/section[1]/div/section[1]/div/dl/div[6]/dd").text
        except:
            founded_vendor = " "
            
        #Specialties - Vendor
        try:
            specialties_vendor = driver. find_element(By.XPATH, "//*[@id='main-content']/section[1]/div/section[1]/div/dl/div[7]/dd").text
        except:
            specialties_vendor = " "
            
        print(f"{vendor_number}, {vendor_name}, {website_vendor}")
        table_list = [vendor_name, vendor_href, about_us_vendor, website_vendor, industry_vendor, company_size_vendor, headquarters_vendor, type_founded, founded_vendor, specialties_vendor]
        templist.append(table_list)
        driver.quit()
        time.sleep(0.5)
    except Exception as e:
        print(e)
        time.sleep(10)
df_temp = pd.DataFrame(templist)
df_temp.to_csv(f"/Users/shagarwal/Documents/LinkedIN/RE_LinkedIn/LinkedIn_Scraping_ProductPages/individual_outputs/LinkedIn_PP_S_Run2_3500_to_7000.csv", index=False)
    
```
