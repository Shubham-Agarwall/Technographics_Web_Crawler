```python
from selenium import webdriver
from selenium.webdriver.firefox.options import Options as FirefoxOptions
from selenium.webdriver.common.by import By
from webdriver_manager.firefox import GeckoDriverManager
from selenium.webdriver.firefox.service import Service
import time
from openpyxl import Workbook
import json
from itertools import zip_longest
import pandas as pd
from selenium.common.exceptions import TimeoutException
#from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.wait import WebDriverWait
from selenium.webdriver.support import expected_conditions as ec
from selenium.common.exceptions import NoSuchElementException
from datetime import datetime
import os
import csv
```

```python
#WebScraping LinkedIn

#Function to click see more
#//button[contains(@class, 'products-serp-layout__see-more t-16')]
def click_more_button():
    click_more = True
    try:
        while click_more:
            time.sleep(1)
            element = driver.find_element_by_xpath(("//button[contains(@class, 'products-serp-layout__see-more t-16')]"))
            if element:
                element.click()
            else:
                click_more = False
    except:
        return None

df_temp = pd.DataFrame()

    
P_S_D_data = pd.read_csv("/Users/shagarwal/Downloads/LinkedIN/linkedin_total_data.csv", header=0)
for index,row in P_S_D_data.iterrows():
    templist = []
    parent_cat = row[0]
    sub_cat = row[1]
    href = row[2]
    print(f"--------{parent_cat}, {sub_cat}, {href}------------")
    try:
        firefox_options = FirefoxOptions()
        driver = webdriver.Firefox(service=Service(GeckoDriverManager().install()))
        driver.get(href)
        driver.maximize_window()
        driver.set_page_load_timeout(30)
        click_more_button()
        all_product_list = []
        all_product_list = driver.find_elements(By.XPATH, "//div[@class='product-serp-card__content']")
        category_products=[]
        for prod in all_product_list:
           
            #prod_name
            try:
                prod_name = prod.find_element(By.XPATH, "./h3").text
            except:
                prod_name = " "
            
            #prod_url
            try:
                prod_url = prod.find_element(By.XPATH, "./h3/a").get_attribute('href')
            except:
                prod_url = " "
            
            #vendor_name
            try:
                vendor_name = prod.find_element(By.XPATH, "./h4/a").text
            except:
                vendor_name = " "
                
            #vendor_url
            try:
                vendor_url = prod.find_element(By.XPATH, "./h4/a").get_attribute('href')
            except:
                vendor_url = " "
            
            #desc
            try:
                desc = prod.find_element(By.XPATH, "./a/p").text
            except:
                desc = " "
            
            print(f"{parent_cat} , {sub_cat} , {href} , {prod_name} , {prod_url}, {vendor_name}, {vendor_url}, {desc}")

            table_list = [parent_cat, sub_cat, href, prod_name, prod_url, vendor_name, vendor_url, desc]
            templist.append(table_list)
    except Exception as e:
        print(e)
        time.sleep(10)
    else:
        df_temp = pd.DataFrame(templist)
        df_temp.to_csv(f"/Users/shagarwal/Downloads/LinkedIN/output(07062022)/{parent_cat}_{sub_cat}.csv", index=False)
        driver.quit()
        time.sleep(5)
        
        
        

```
