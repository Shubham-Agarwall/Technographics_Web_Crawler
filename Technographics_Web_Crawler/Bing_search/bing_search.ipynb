```python
#Bing Search : Step 2

df_temp = pd.DataFrame()

#Reading sample input
input_products = pd.read_csv("/Users/shagarwal/Documents/Bing_search/input_sheet.csv")
for index,row in input_products.iterrows():
    templist = []
    keywords = row[0]
    print(keywords)
    #Going to the webpage
    try:
        options = webdriver.ChromeOptions()
        ua = UserAgent()
        userAgent = ua.random
        print(userAgent)
        options.add_argument(f'user-agent={userAgent}' )
        driver = uc.Chrome(
            options=options,
        ) 
        driver.maximize_window()
        driver.get("https://www.bing.com/search?q=%22&form=QBLH&sp=-1&pq=%22&sc=9-1&qs=n&sk=&cvid=A8FB92972727456D88F03F75F447ADA7&ghsh=0&ghacc=0&ghpl=")
        driver.set_page_load_timeout(30)
        
        try:
            #//*[@id="sb_form_q"]
            driver.find_element(By.XPATH, "//*[@id='sb_form_q']").click()
            driver.find_element(By.XPATH, "//*[@id='sw_clx']").click()
        except:
            print("could not clear the search bar")
    except Exception as e:
        print(e)
        time.sleep(10)
        break
    #Enter product name and click on search button
    try:
        driver.find_element(By.XPATH, "//*[@id='sb_form_q']").send_keys(keywords)
        driver.find_element(By.XPATH, "//*[@id='sb_form_go']").click()
    except:
        print("could not perform google search")
    
    #Get the url
        #//*[@id="rso"]/div[1]/div/div/div/div[1]/div/div/div[1]/div/a
        #//*[@id="rso"]/div[1]/div/div/div[1]/div/a
        #//*[@id='rso']/div[1]/div/div/div/div[1]/div/a
        #//*[@id='rso']/div[1]/div/div/div[1]/div/div/div[1]/div/a
    try:
        product_url_one = driver.find_element(By.XPATH, "//*/li[1]/div[1]/div/h2/a").get_attribute('href')
    except:
        product_url_one = " "
        #print("product_url_one Not Found")
    try:
        product_url_two = driver.find_element(By.XPATH, "//*/li[1]/h2/a").get_attribute('href')
    except:
        product_url_two = " "
        #print("product_url_two Not Found")
    try:
        product_url_three = driver.find_element(By.XPATH, "//*/li[2]/h2/a").get_attribute('href')
    except:
        product_url_three = " "
        #print("product_url Not Found")
    try:
        product_url_four = driver.find_element(By.XPATH, "//*[@id='b_results]/li[1]/div[1]/div/h2/a").get_attribute('href')
    except:
        product_url_four = " "
        #print("product_url_four Not Found")
        
    #get the the whole page content in text
    try: 
        if len(product_url_one) > 1:
            driver.find_element(By.XPATH, "//*/li[1]/div[1]/div/h2/a").click()
            driver.set_page_load_timeout(10)
            content_url_one = driver.find_element(By.XPATH, "/html/body").text
            driver.execute_script("window.history.go(-1)")
            #driver.back()
        else:
            content_url_one = "none1"
    except:
        print("could not get any content from product_url_one")
        
    try: 
        if len(product_url_two) > 1:
            driver.find_element(By.XPATH, "//*/li[1]/h2/a").click()
            driver.set_page_load_timeout(10)
            content_url_two = driver.find_element(By.XPATH, "/html/body").text
            driver.execute_script("window.history.go(-1)")
        else:
            content_url_two = "none2"
    except:
        print("could not get any content from product_url_two")
        
    try: 
        if len(product_url_three) > 1:
            driver.find_element(By.XPATH, "//*/li[2]/h2/a").click()
            driver.set_page_load_timeout(10)
            content_url_three = driver.find_element(By.XPATH, "/html/body").text
            driver.execute_script("window.history.go(-1)")
        else:
            content_url_three = "none3"
    except:
        print("could not get any content from product_url_three")
        
    try: 
        if len(product_url_four) > 1:
            driver.find_element(By.XPATH, "//*[@id='b_results]/li[1]/div[1]/div/h2/a").click()
            driver.set_page_load_timeout(10)
            content_url_four = driver.find_element(By.XPATH, "/html/body").text
            driver.execute_script("window.history.go(-1)")
        else:
            content_url_four = "none4"
    except:
        print("could not get any content from product_url_four")
    
    print(f"{keywords},{product_url_one}, {content_url_one}, {product_url_two}, {content_url_two} , {product_url_three}, {content_url_three} {product_url_four}, {content_url_four} ")
    table_dict = {'keywords': keywords, 'product_url_one' : product_url_one, 'content_url_one' : content_url_one, 'product_url_two' : product_url_two, 'content_url_two' : content_url_two, 'product_url_three' : product_url_three, 'content_url_three' : content_url_three, 'product_url_four' : product_url_four, 'content_url_four' : content_url_four }
    templist.append(table_dict)
    df_temp = pd.DataFrame(templist)
    df_temp.to_csv(f"/Users/shagarwal/Documents/Bing_search/{keywords}.csv", index=False)
    time.sleep(4)
    driver.quit()
    
```
