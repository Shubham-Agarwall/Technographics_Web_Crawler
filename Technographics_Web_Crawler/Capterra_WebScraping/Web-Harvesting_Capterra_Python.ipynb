```python
from selenium import webdriver
from selenium.webdriver.firefox.options import Options as FirefoxOptions
#from fake_useragent import UserAgent
from selenium.webdriver.common.action_chains import ActionChains
#pip install geckodriver
from webdriver_manager.firefox import GeckoDriverManager
from selenium.webdriver.firefox.service import Service
from selenium.webdriver.common.by import By
from selenium.common.exceptions import TimeoutException
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.wait import WebDriverWait
from selenium.common.exceptions import NoSuchElementException
from datetime import datetime
import time
import os
import pandas as pd
import csv
import json
```

```python
firefox_options = FirefoxOptions()
driver = webdriver.Firefox(service=Service(GeckoDriverManager().install()))
#driver = webdriver.Firefox(Firefox)
driver.get('https://www.capterra.com/categories')
driver.maximize_window()
driver.set_page_load_timeout(30)


parent_Sub_Categories_list = []
parent_category_list = []
templist = []

#//div/div/div/div/div[contains(@class, 'browse-group')]
parent_Sub_Categories_list = driver.find_elements(By.XPATH, "//div/div/div/div/div[contains(@class, 'browse-group')]")
for element in parent_Sub_Categories_list:
    #//div[contains(@class, 'browse-alpha')]
    Parent_cat_number = element.find_element(By.XPATH, ".//div[contains(@class, 'browse-alpha')]").text
    #//li/a
    parent_category_list = element.find_elements(By.XPATH, ".//li/a")
    for ele in parent_category_list:
        parent_category = ele.text
        parent_category_href = element.find_element(By.XPATH, ".//ol/li/a").get_attribute('href')
        print(f"{Parent_cat_number}, {parent_category}, {parent_category_href}")
        
        table_dict = {'Parent_cat_number': Parent_cat_number, 'parent_category': parent_category, 'domain': parent_category_href}
        templist.append(table_dict)
        df = pd.DataFrame(templist)
df.to_csv('Capterra_V1.csv')

    
    

driver.quit()
```

```python
#Click "Show More" until there is no show more button
def click_more_button():
    click_more = True
    try:
        while click_more:
            time.sleep(1)
            #//div[@class='nb-mx-3xs' and text()='Show More']
            element = driver.find_element_by_xpath("//div[@class='nb-mx-3xs' and text()='Show More']")
            if element:
                element.click()
            else:
                click_more = False
    except:
        return None

    

def get_element_data(element):
    description = ""
    try:
        description = element.text
    except Exception as e:
        return description
    finally: 
        return description
    
    
df_temp = pd.DataFrame()

parent_Sub_Categories_from_csv = pd.read_csv('/Users/shagarwal/Documents/Capterra_WebScraping/input/capterra_input.csv', header=0)
for index,row in parent_Sub_Categories_from_csv.iterrows():
    templist = []
    main_category_no = row[1]
    main_category = row[2]
    main_category_href = row[3]
    print(f"*****************:> {main_category_no}, {main_category} <:*****************")
    try:
        firefox_options = FirefoxOptions()
        driver = webdriver.Firefox(service=Service(GeckoDriverManager().install()))
        driver.get(main_category_href)
        driver.maximize_window()
        driver.set_page_load_timeout(30)
        click_more_button()
        #All rows in the product page : //div[@class = 'nb-block nb-w-100']
        all_row_data = []
        all_row_data = driver.find_elements(By.XPATH, "//div[@class = 'nb-block nb-w-100' and @data-testid]")
        #print(f"*****************:> {main_category_no}, {main_category} <:*****************")
        for row_data in all_row_data:
            #print(row_data.text)
            #Product_name
            try:
                #product name : 
                #//div[@class = 'nb-block nb-w-100']//div/h2/a
                product_name = get_element_data(row_data.find_element(By.XPATH, ".//div/div//h2/a"))
                #print(product_name)
            except:
                product_name = " "

            #Product_href
            try:
                #product_href : 
                #.//div/div//h2/a
                product_href = row_data.find_element(By.XPATH, ".//div/div//h2/a").get_attribute('href')
                #print(product_href)
            except:
                product_href = " "

            #Vendor : 
            #//div[@class = 'nb-block nb-w-100' and @data-testid]//div/h3/span
            try:
                vendor_name = get_element_data(row_data.find_element(By.XPATH, ".//div/h3/span"))
                exact_vendor_name = vendor_name[3:]
                #print(vendor_name[3:])
            except:
                vendor_name = " "

            #Pr0duct_rating : 
            #//div[@class = 'nb-block nb-w-100' and @data-testid]//div/div/div/a/div/div[@class = 'nb-ml-3xs' ]
            try:
                product_rating = get_element_data(row_data.find_element(By.XPATH, ".//div/div/div/a/div/div[@class = 'nb-ml-3xs']"))
                #print(product_rating)
            except:
                product_rating = " "

            #Product_VisitWebsite_URL:
            #//div[@class = 'nb-block nb-w-100' and @data-testid]//div/a[@class = 'nb-button nb-button-standard nb-button-preferred nb-w-full sm:nb-w-auto']
            try:
                product_vw_url = row_data.find_element(By.XPATH, ".//div/a[@class = 'nb-button nb-button-standard nb-button-preferred nb-w-full sm:nb-w-auto']").get_attribute('href')
                #print(product_vw_url)
            except:
                product_vw_url = " "

            #Product_View_Profile_URL: //div[@class = 'nb-block nb-w-100' and @data-testid]//div/div/div/div/a[@class = 'nb-button nb-button-standard nb-button-primary ']
            try:
                view_profile_url = row_data.find_element(By.XPATH, ".//div/div/div/div/a[@class = 'nb-button nb-button-standard nb-button-primary ']").get_attribute('href')
                #print(view_profile_url)
            except:
                view_profile_url = " "

            #product Description: 
            #//div[@class = 'nb-block nb-w-100' and @data-testid]//div/div[@class = 'nb-w-full nb-hidden sm:nb-block']
            try:
                p_d = get_element_data(row_data.find_element(By.XPATH, ".//div/div[@class = 'nb-w-full nb-hidden sm:nb-block']"))
                #print(p_d)
            except:
                p_d = " "

            # EMERGING FAVORITE/TOP PERFORMER/ESTABLISHED PLAYER/NOTEWORTHY PRODUCT(button): 
            #.//div/div/button/div/div
            try:
                special_comment = get_element_data(row_data.find_element(By.XPATH, ".//div/div/button/div/div"))
                #print(emerging_favorite)
            except:
                special_comment = " "

            print(f"{product_name} , {product_href} , {exact_vendor_name} , {product_rating} , {product_vw_url }, {view_profile_url}, {p_d}, {special_comment}")
            print("--x--x--x--x--x--x--x--x--x--x--x--x--x--x--x--x--x--x--x--x--x--x--x--x--x--x--x--x--x--x--")
            table_dict = {'main_category_no': main_category_no, 'main_category': main_category, 'main_category_href': main_category_href, 'product_name': product_name, 'product_href': product_href, 'vendor_name': exact_vendor_name, 'product_rating': product_rating, 'product_visitwebsite_url': product_vw_url, 'view_profile_url': view_profile_url, 'product_description': p_d, 'special_comment': special_comment,}
            templist.append(table_dict)
    except Exception as e:
        print(e)
        time.sleep(10)
    else:
        df_temp = pd.DataFrame(templist)
        df_temp.to_csv(f"/Users/shagarwal/Documents/Capterra_WebScraping/output/{main_category_no}_{main_category}.csv", index=False)
        driver.quit()
        time.sleep(5)
        

```
