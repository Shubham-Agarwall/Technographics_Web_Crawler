```python
import wikipediaapi
import pandas as pd
import csv
try:
    from googlesearch import search
except ImportError:
    print("No module named 'google' found")
```

```python
#Finding Wikipedia link for Product
templist = []
product_data = pd.read_csv("/Users/shagarwal/wikipedia/ProductWiki.csv", header=0)
for index,row in product_data.iterrows():

    wiki_product = row[0]
    parent_category = row[1]
    sub_category = row[2]
    print("-------------->",wiki_product ,"<------------------")

    # to search
    query = wiki_product
    
    #for search method refer imports googlesearch:
    for j in search(query, tld="com", num=10, stop=10, pause=2 ,country='US'):
        if "en.wikipedia.org" in j:
            good_wiki_domain = j
            print(f"{wiki_product}, {parent_category}, {sub_category}, {good_wiki_domain}")
            
            table_dict = {'Product': wiki_product, 'Parent Category': parent_category, 'Sub Category': sub_category, 'Wiki Domain': good_wiki_domain}
            templist.append(table_dict)
            df = pd.DataFrame(templist)
df.to_csv("Wiki_P_PC_SC_WD.csv")

```
