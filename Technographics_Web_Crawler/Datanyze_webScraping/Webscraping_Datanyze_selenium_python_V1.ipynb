# Basics of Selenium 

```python
# Starting a session.
driver = webdriver.Chrome()

#Take action on browser, navigating to webpage.
#driver.get() : It's used to go to the particular website , But it doesn't maintain the browser History and cookies so , we can't use forward and backward button , if we click on that , page will not get schedule
driver.get("http://www.google.com")
#driver.navigate() : it's used to go to the particular website , but it maintains the browser history and cookies, so we can use forward and backward button to navigate between the pages during the coding of Testcase
driver.navigate().to(<type_url>) #or driver.navigate(<type_url>)
driver.back()
driver.forward()
driver.refresh()


#Requesting browser information
driver.title
driver.current_url

#Establish waiting strategy/create delay
#1.Implicite wait- waits for the mentioned time in seconds. 
driver.implicitely_wait(10) #Implicite wait- waits for the mentioned time in seconds. 

#2.Explicit wait-waits for certain condition or for maximum time mentioned it is an intelligent wait #list_of_conditions= "https://www.selenium.dev/selenium/docs/api/py/webdriver_support/selenium.webdriver.support.expected_conditions.html?highlight=expected"
WebDriverWait(driver, timeout=3).until(some_condition) 

#3.The Fluent Wait in Selenium is used to define maximum time for the web driver to wait for a condition, as well as the frequency with which we want to check the condition before throwing an “ElementNotVisibleException” exception. It checks for the web element at regular intervals until the object is found or timeout happens.
# Frequency: Setting up a repeat cycle with the time frame to verify/check the condition at the regular interval of time
#fluent wait is the ideal wait to use as this will try to find the element at different frequency until it finds it or the final timer runs out.
WebDriverWait(driver, timeout=10, poll_frequency=1, ignored_exceptions=[ElementNotVisibleException, ElementNotSelectableException])


#Finding an element
search_box = driver.find_element(By.NAME, "q")
search_button = driver.find_element(By.NAME, "btnK")


#Take action on element
search_box.send_keys("Selenium")
search_button.click()


#Request element information
driver.find_element(By.NAME, "q").get_attribute("value") # => "Selenium"


#End the session
driver.quit()

```

# Below is the function to initilise the webdriver and add options such as headless("--headless")
# Headless browsers can work without displaying any graphical UI, which allows applications to be a single source of interaction for users and provides a smooth user experience. Selenium helps you make any browser headless by adding an options argument as --headless.


```python
#function to configure chrome webdriver

def configure_chrome_driver():
    #adding additional options to the driver
    chrome_options = ChromeOptions()
    #Adding the argument --headless using .add_argument attribute on the object ChromeOptions() created above
    ChromeOptions.add_argument("--headless")
    #Now initialising the webdriver by providing EXECUTABLE-PATH to the chrome-webdriver-downloaded and copied to /usr/local/bin/chromedriver
    #If chromedriver is in PATH(i.e, bin folder), need not provide full EXECUTABLE-PATH
    chrome_driver_path = "/usr/local/bin/chromedriver"
    driver = webdriver.Chrome(executable_path = chrome_driver_path, options = chrome_options)
    return driver
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager
import time
from openpyxl import Workbook
import json
from itertools import zip_longest
import pandas as pd
```

```python
driver = webdriver.Chrome(ChromeDriverManager().install())
driver.maximize_window()
driver.get("https://www.datanyze.com/market-share")
driver.implicitly_wait(10)

parent_category_list = []
sub_category_list = []
sub_category_href_list = []
tech_count_sub_category_list = []

#Parent Category
#for parent and sub category- //div[contains(@class,'industry-title-wrapper')]
#for just parent category-  //div[contains(@class,"industry-title-wrapper
parent_categories = driver.find_elements(By.XPATH ,'//div[contains(@class,"industry-title-wrapper")]')
for each_p_c in parent_categories:
    #print(each_p_c.text)
    parent_category_list.append(each_p_c.text)
    #parent_cat = each_p_c.find_element(By.TAG_NAME,'div').text
    


#Sub category   
# //a[contains(@class,"category-name")]
sub_categories = driver.find_elements(By.XPATH,'//a[contains(@class,"category-name")]')
for each_s_c in sub_categories:
    #print(each_s_c.text)
    sub_category_list.append(each_s_c.text)
    
#No of technologies in each sub category
#//span[contains(@class, 'category-number')]
tech_count_sub_category = driver.find_elements(By.XPATH,'//span[contains(@class,"category-number")]')
for each_tech_count in tech_count_sub_category:
    tech_count_sub_category_list.append(each_tech_count.text)
    

#links for each sub-category(href)
#	Use get_attribute('href) – save as url
#//a[contains(@href, '/market-share/')]
sub_category_href = driver.find_elements(By.XPATH,'//a[contains(@href, "/market-share/")]')
for each_sub_cat_href in sub_category_href:
    #len(sub_category_href)
    #print(each_sub_cat_href.get_attribute("href"))
    #sub_cat_href = each_sub_cat_href.get_attribute("href")
    sub_category_href_list.append(each_sub_cat_href.get_attribute("href"))

    
P_SP_D_categories = []

P_SP_D_categories = driver.find_elements(By.XPATH, "//li/a")
for element in P_SP_D_categories:
    sub_cat = element.text
    href = element.get_attribute('href')
    parent_cat = element.find_element(By.XPATH, './/parent::li//parent::ul//preceding-sibling::div').text
    #print(f"{parent_cat}, {sub_cat}, {href}")
    
#Parsing sub-categories, one by one.
driver.find_element(By.XPATH,"//a[text()='Accounting']").click()





driver.quit()
```

```python
#use the code in next shell.

#zip all data and export to xlsx
data_list = zip(sub_category_list, sub_category_href_list, tech_count_sub_category_list)

wb = Workbook()
sheet_xlsx = wb.active

for data in list(data_list):
    sheet_xlsx.append(data)
wb.save("Datanyze_data.csv")



# sheet.cell(row=Count, column=6).value = resultado_pesquisa.text 
```
