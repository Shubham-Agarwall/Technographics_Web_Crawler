```python
#Enrich already scraped TrustRadius product catalog.

os.environ['GH_TOKEN'] = "removed-for-github-upload"

df_temp = pd.DataFrame()

data = pd.read_csv("/Users/shagarwal/Documents/WebHarvesting_TrustRadius/Enrich_TR/Input/TR_Products.csv")
for index,row in data[:25].iterrows():
    templist = []
    product_name = row[0]
    href = row[1]
    #print(f"--------{parent_cat}, {sub_cat}, {href}------------")
    try:
        firefox_options = FirefoxOptions()
        firefox_options.add_argument("--headless")
        driver = webdriver.Firefox(options=firefox_options, service=Service(GeckoDriverManager().install()))
        driver.maximize_window()
        driver.get(href)
        driver.set_page_load_timeout(30)
        time.sleep(2)
        
        #1
        #//*[@id='react_buyer_app']/div/header/div/div/ul/li/a"
        try:
            sub_category = driver.find_element(By.XPATH, "//*[@id='react_buyer_app']/div/header/div/div/ul/li/a").text
        except:
            sub_category = " "
            
        #2
        #//*[@id="react_buyer_app"]/div/header/div/div/div/h1/div/text()[2]
        try:
            formerly = driver.find_element(By.XPATH, "//*[@class='ProductPageHeader_former-product-name__36Pzc']").text[9:]
        except:
            formerly = " "
        
        #3
        #//*[@id="react_buyer_app"]/div/header/div/div[2]/div/div/div/div/div/span
        try:
            top_rated = driver.find_element(By.XPATH, "//*[@id='react_buyer_app']//div/header/div/div[2]/div/div/div/div/div/span").text
        except:
            top_rated = " "
        
        #4
        #//*[@id="react_buyer_app"]/div/header/div/div/ul/li/button/text()[1]
        try:
            reviews_ratings = driver.find_element(By.XPATH, "//*[@id='react_buyer_app']/div/header/div/div/ul/li/button").text
        except:
            reviews_ratings = " "
        
        #5
        #//*[@id="react_buyer_app"]/div/header/div/div/div/div/div/div/span[2]
        try:
            score = driver.find_element(By.XPATH, "//*[@id='react_buyer_app']/div/header/div/div/div/div/div/div/span[2]").text
        except:
            score = " "
            
        #6
        # //*[@id="rail-id"]/div[2]/div/a/span
        try:
            button_1_name = driver.find_element(By.XPATH, "//*[@id='rail-id']/div[2]/div/a/span").text
        except:
            button_1_name = " "
            
        #7.1
        #
        try:
            button_1_url = driver.find_element(By.XPATH, "//*[@class= 'ProductCTASection_cta-button___PqFI button button-secondary-pill']").get_attribute('href')
        except:
            button_1_url = " "
        
        #8
        #
        try:
            vendor_name = driver.find_element(By.XPATH, "//*[@id='rail-id']/aside/h2/a").text
        except:
            vendor_name = " "
            
        #9
        #//*[@id="rail-id"]/aside/h2/a
        try:
            vendor_href = driver.find_element(By.XPATH, "//*[@id='rail-id']/aside/h2/a").get_attribute('href')
        except:
            vendor_href = " "
        
        #10
        #
        try:
            vendor_info = driver.find_element(By.XPATH, "//*[@id='react_buyer_app']/script[1]").get_attribute('innerHTML')
        except:
            vendor_info = " "
            
        time.sleep(1)
        #print(f" {sub_category}, {formerly}, {top_rated}, {reviews_ratings}, {score}, {button_1_name}, {button_1_url}, {vendor_name} ")
        print(f" {index}, {product_name}, {sub_category} ")
        table_dict = {'product_name': product_name, 'href_TR': href, 'sub_category': sub_category, 'formerly': formerly, 'top_rated' : top_rated, 'reviews_ratings' : reviews_ratings, 'score' : score, 'button_1_name' : button_1_name, 'button_1_url' : button_1_url,'vendor_name': vendor_name , 'vendor_href':vendor_href , 'vendor_info' : vendor_info }
        templist.append(table_dict)
        df_temp = pd.DataFrame(templist)
        df_temp.to_csv(f"/Users/shagarwal/Documents/WebHarvesting_TrustRadius/Enrich_TR/output/Individual_output/{index}_{product_name}.csv", index=False)
        driver.quit()
        time.sleep(1)
    except Exception as e:
        print(e)
        time.sleep(10)       
   
```
