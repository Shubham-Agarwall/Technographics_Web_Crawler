```python
#Bing Search

df_temp = pd.DataFrame()

#Reading sample input
input_products = pd.read_csv("/Users/shagarwal/Documents/HG_Insights/Bing_Product_URLs_5k_Products/input/hginsight_nonmatched_vendor_with_demandbase_catalog_recent.csv")
for index,row in input_products[1500:].iterrows():
    templist = []
    product_name = row[2]
    #print(keywords)
    #Going to the webpage
    try:
        chrome_options = webdriver.ChromeOptions()
        #adding functionality to open chrome in incognito using "--incognito"
        chrome_options.add_argument("--incognito")
        chrome_options.add_argument("headless")
        driver = webdriver.Chrome(ChromeDriverManager().install(), options = chrome_options)
        driver.maximize_window()
        driver.get("https://www.bing.com/search?q=%22&form=QBLH&sp=-1&pq=%22&sc=9-1&qs=n&sk=&cvid=A8FB92972727456D88F03F75F447ADA7&ghsh=0&ghacc=0&ghpl=")
        driver.set_page_load_timeout(30)
    
        try:
            #//*[@id="sb_form_q"]
            time.sleep(0.5)
            driver.find_element(By.XPATH, "//*[@id='sb_form_q']").click()
            time.sleep(0.5)
            driver.find_element(By.XPATH, "//*[@id='sw_clx']").click()
        except:
            print("could not clear the search bar")
            
    
        #Enter product name and click on search button
        try:
            driver.find_element(By.XPATH, "//*[@id='sb_form_q']").send_keys(product_name)
            driver.find_element(By.XPATH, "//*[@id='sb_form_q']").send_keys(" ")
            driver.find_element(By.XPATH, "//*[@id='sb_form_q']").send_keys("Product URL")
            driver.find_element(By.XPATH, "//*[@id='sb_form_go']").click()
        except:
            print("could not perform google search")

        #Get the url
        #//*[@id='b_results']/li[1]/div[1]/h2/a
        #//*[@id='b_results']/li[1]/div[1]/div[1]/h2/a
        #//*[@id='b_results']/li[2]/h2/a
        #//*[@id='b_results']/li[1]/h2/a
        #//*[@id="b_results"]/li[1]/h2/a
        try:
            product_url_1 = driver.find_element(By.XPATH, "//*[@id='b_results']/li[1]/div[1]/h2/a").get_attribute('href')
        except:
            product_url_1 = " "
            #print("product_url_one Not Found")
        try:
            product_url_2 = driver.find_element(By.XPATH, "//*[@id='b_results']/li[1]/div[1]/div[1]/h2/a").get_attribute('href')
        except:
            product_url_2 = " "
            #print("product_url_two Not Found")
        try:
            product_url_3 = driver.find_element(By.XPATH, "//*[@id='b_results']/li[2]/h2/a").get_attribute('href')
        except:
            product_url_3 = " "
            #print("product_url Not Found")
        try:
            product_url_4 = driver.find_element(By.XPATH, "//*[@id='b_results']/li[1]/h2/a").get_attribute('href')
        except:
            product_url_4 = " "
            #print("product_url_four Not Found")

        print(f"{index}, {product_name}, {product_url_1}, {product_url_2}, {product_url_3}, {product_url_4}")
        table_dict = {'product_name': product_name, 'product_url_1' : product_url_1, 'product_url_2' : product_url_2, 'product_url_3' : product_url_3, 'product_url_4' : product_url_4 }
        templist.append(table_dict)
        df_temp = pd.DataFrame(templist)
        df_temp.to_csv(f"/Users/shagarwal/Documents/HG_Insights/Bing_Product_URLs_5k_Products/Individual_output/{product_name}.csv", index=False)
        time.sleep(1)
        driver.quit()
    
    except Exception as e:
        print(e)
        time.sleep(10)
        break
```
